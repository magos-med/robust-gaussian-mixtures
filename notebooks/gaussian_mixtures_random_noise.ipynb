```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
import plotly.express as px
import plotly.io as pio

from clustering import (StandardGaussianMixture, SpearmanGaussianMixture, MADSpearmanGaussianMixture)
from utils import plot_gaussian_mixtures

pio.templates.default = 'ggplot2'
pio.renderers.default = "svg"
```

```python
# Create a dataset of 2D distributions
n_clusters = 3
x_train, true_labels = make_blobs(n_samples=100, centers=n_clusters, random_state=42, center_box=[-10, 10])
x_train_scaled = StandardScaler().fit_transform(x_train)
```

```python
px.scatter(x=x_train_scaled[:, 0], y=x_train_scaled[:, 1], color=true_labels)
```

```python
gmm = StandardGaussianMixture(n_components=n_clusters)
gmm.fit(x_train_scaled)
gmm.accuracy_score(x_train_scaled, true_labels)
```

```python
plot_gaussian_mixtures(gmm, x_train_scaled)
```

```python
gmm = MADSpearmanGaussianMixture(n_components=n_clusters, max_iter=100)
gmm.fit(x_train_scaled)
gmm.accuracy_score(x_train_scaled, true_labels)
```

```python
plot_gaussian_mixtures(gmm, x_train_scaled)
```

```python
noise = (np.random.random((5, 2)) - 0.5) * 20
noise_labels = np.random.randint(0, 3, 5)
```

```python
x_train = np.append(x_train, noise, axis=0)
true_labels = np.append(true_labels, noise_labels)
x_train_scaled = StandardScaler().fit_transform(x_train)
```

```python
px.scatter(x=x_train_scaled[:, 0], y=x_train_scaled[:, 1], color=true_labels)
```

```python
gmm = StandardGaussianMixture(n_components=n_clusters)
gmm.fit(x_train_scaled)
gmm.accuracy_score(x_train_scaled, true_labels)
```

```python
plot_gaussian_mixtures(gmm, x_train_scaled)
```

```python
gmm = SpearmanGaussianMixture(n_components=n_clusters, max_iter=100)
gmm.fit(x_train_scaled)
gmm.accuracy_score(x_train_scaled, true_labels)
```

```python
plot_gaussian_mixtures(gmm, x_train_scaled)
```

```python
gmm = MADSpearmanGaussianMixture(n_components=n_clusters, max_iter=100)
gmm.fit(x_train_scaled)
gmm.accuracy_score(x_train_scaled, true_labels)
```

```python
plot_gaussian_mixtures(gmm, x_train_scaled)
```
