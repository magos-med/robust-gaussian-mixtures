```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
import plotly.io as pio

from clustering import (
    StandardGaussianMixture,
    SpearmanGaussianMixture, MADSpearmanGaussianMixture,
    KendallGaussianMixture, MADKendallGaussianMixture,
    OrtizGaussianMixture, MADOrtizGaussianMixture,
    ApproxOrtizGaussianMixture, MADApproxOrtizGaussianMixture,
    ChoquetGaussianMixture, OrtizChoquetGaussianMixture, ApproxOrtizChoquetGaussianMixture
)
from utils import plot_gaussian_mixtures

pio.templates.default = 'ggplot2'
pio.renderers.default = "svg"
```

```python
# Create a dataset of 2D distributions
n_clusters = 3
x_train, true_labels = make_blobs(n_samples=100, centers=n_clusters, random_state=42, center_box=[-10, 10])
```

```python
noise, noise_labels = make_blobs(n_samples=2, centers=[[20, 1]])
x_train = np.append(x_train, noise, axis=0)
true_labels = np.append(true_labels, noise_labels)
x_train_scaled = StandardScaler().fit_transform(x_train)
```

```python
gmm = StandardGaussianMixture(n_components=n_clusters)
gmm.fit(x_train_scaled)
gmm.accuracy_score(x_train_scaled, true_labels)
```

```python
plot_gaussian_mixtures(gmm, x_train_scaled)
```

```python
gmm = SpearmanGaussianMixture(n_components=n_clusters)
gmm.fit(x_train_scaled)
gmm.accuracy_score(x_train_scaled, true_labels)
```

```python
plot_gaussian_mixtures(gmm, x_train_scaled)
```
